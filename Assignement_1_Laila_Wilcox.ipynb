```python
# Links Used
# https://leetcode.com/problems/valid-palindrome/solutions/7197820/clean-simple-efficient-beginner-friendly-easy-explanation/
# https://leetcode.com/problems/valid-palindrome/solutions/7219641/the-two-pointer-dance-a-classic-o-1-space-solution-for-valid-palindrome/
# https://leetcode.com/problems/valid-palindrome/solutions/6819917/beats-100-easiest-solution-for-beginners-c-java-python-javascript/

class Solution:
    def isPalindrome(self, s: str) -> bool:
        left = 0
        right = len(s) - 1

        while left < right:

            while left < right and not s[left].isalnum():
                left += 1

            while left < right and not s[right].isalnum():
                right -= 1

            if s[left].lower() != s[right].lower():
                return False

            left += 1
            right -= 1

        return True
```

## 3. Final Correct Code



```python
# AI Disclaimer: I used ClaudeAI to see how good my draft code was and ended up simplifying it to this.
# AI Disclaimer: I also used ClaudeAI to understand specific lines of code such as .isalnum(), that I am unfamiliar with

class Solution:
    def isPalindrome(self, s: str) -> bool:
        chars = []
        for c in s:
            if c.isalnum():
                chars.append(c.lower())
        candidate = ''.join(chars)
        return candidate == candidate[::-1]
```

```python
solution = Solution()
result = solution.isPalindrome("A man, a plan, a canal: Panama")
print(result)
```
